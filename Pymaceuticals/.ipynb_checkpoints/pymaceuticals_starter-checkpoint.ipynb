## Observations and Insights 

```python
# Observations: 
# The Ramicane method seems to be the most effective method overall. Since it has the lowest mean, 
# and the best quartile data out of the 4 best regimens.
# The data seems to have a strong uphill (positive) linear relationship since it's above .7. So, studies are consistent
# The tumor volumes on average decrease a 48.8%, so one could say the experiment seems pretty reliable



# Dependencies and Setup
import matplotlib.pyplot as plt
import pandas as pd
import scipy.stats as st


# Study data files
mouse_metadata_path = "data/Mouse_metadata.csv"
study_results_path = "data/Study_results.csv"

# Read the mouse data and the study results
mouse_metadata = pd.read_csv(mouse_metadata_path)
study_results = pd.read_csv(study_results_path)

# Combine the data into a single dataset

df_merged = pd.merge(mouse_metadata, study_results, on="Mouse ID", how="left")

# Display the data table for preview

df_merged.head()
```

```python
# Checking the number of mice.
print(f'There are {len(df_merged)} mice')
```

## Summary Statistics

```python
# Generate a summary statistics table of mean, median, variance, standard deviation, and SEM of the tumor volume for each regimen
# Use groupby and summary statistical methods to calculate the following properties of each drug regimen: 
# mean, median, variance, standard deviation, and SEM of the tumor volume. 
# Assemble the resulting series into a single summary dataframe.

# Calculations

mean_df = df_merged.groupby('Drug Regimen')['Tumor Volume (mm3)'].mean()
median_df= df_merged.groupby('Drug Regimen')['Tumor Volume (mm3)'].median()
variance_df= df_merged.groupby('Drug Regimen')['Tumor Volume (mm3)'].var()
std_df = df_merged.groupby('Drug Regimen')['Tumor Volume (mm3)'].std()
sem_df = df_merged.groupby('Drug Regimen')['Tumor Volume (mm3)'].sem()

# Data Frame

summary_table = pd.DataFrame({'Mean (mm3)': mean_df, 'Median (mm3)': median_df, 'Variance': variance_df, 'Std': std_df, 'SEM': sem_df})


# Display Table

summary_table

```

```python
# Generate a summary statistics table of mean, median, variance, standard deviation, and SEM of the tumor volume for each regimen
# Using the aggregation method, produce the same summary statistics in a single line

summary_aggregation = df_merged.groupby('Drug Regimen')['Tumor Volume (mm3)'].agg(['mean', 'median', 'var','std', 'sem'])
summary_aggregation
```

## Quartiles, Outliers and Boxplots

```python
# Calculate the final tumor volume of each mouse across four of the treatment regimens:  
# Capomulin, Ramicane, Infubinol, and Ceftamin
# Start by getting the last (greatest) timepoint for each mouse
# Merge this group df with the original dataframe to get the tumor volume at the last timepoint


# Since the last point for the mouses is 45, I only took those measures and the duplicate mouse id's 
# get deleted automatically. Then I took only the ones that where under the 4 regimens

greatest_tp = df_merged.loc[df_merged['Timepoint'] == 45]

df_capomulin = greatest_tp[greatest_tp['Drug Regimen'].isin(['Capomulin'])]
df_capomulin = df_capomulin.sort_values(["Tumor Volume (mm3)"], ascending=True).reset_index()

df_ramicane = greatest_tp[greatest_tp['Drug Regimen'].isin(['Ramicane'])]
df_ramicane = df_ramicane.sort_values(["Tumor Volume (mm3)"], ascending=True).reset_index()

df_infubinol = greatest_tp[greatest_tp['Drug Regimen'].isin(['Infubinol'])]
df_infubinol = df_infubinol.sort_values(["Tumor Volume (mm3)"], ascending=True).reset_index()

df_ceftamin = greatest_tp[greatest_tp['Drug Regimen'].isin(['Ceftamin'])]
df_ceftamin = df_ceftamin.sort_values(["Tumor Volume (mm3)"], ascending=True).reset_index()

df_capomulin
```

```python
from scipy.stats import iqr
import numpy as np

# Put treatments into a list for for loop (and later for plot labels)
# I didn't use this for my procedure

# Create empty list to fill with tumor vol data (for plotting)
# Capomulin 

tumor_vol = []
tumor_vol.append(df_capomulin['Tumor Volume (mm3)'])

# Calculate the IQR and quantitatively determine if there are any potential outliers. 
# Locate the rows which contain mice on each drug and get the tumor volumes

lower_quar =  np.percentile(tumor_vol,25)
upper_quar = np.percentile(tumor_vol, 75)
iqr = iqr(tumor_vol)
print(f'The lower quartile is {lower_quar}')
print(f'The uper quartile is {upper_quar}')
print(f'The IQR quartile is {iqr}')

 
# add subset 
# Determine outliers using upper and lower bounds

upper_bound = upper_quar + (1.5*iqr)
lower_bound = lower_quar - (1.5*iqr)

print(f'The Values over {upper_bound} and below {lower_bound} could be outliers.')

    
    
```

```python
# Generate a box plot of the final tumor volume of each mouse across four regimens of interest



fig1, ax1 = plt.subplots()
ax1.set_title('Final Tumor Volume for the Capomulin Regimen')
ax1.set_ylabel('Volume (mm3)')
green_diamond = dict(markerfacecolor='g', marker='D')
ax1.boxplot(tumor_vol, flierprops= green_diamond)
plt.show
```

## Line and Scatter Plots

```python
# Generate a line plot of tumor volume vs. time point for a mouse treated with Capomulin

linep_capomulin = df_merged.loc[df_merged['Drug Regimen'] == 'Capomulin']
linep_capomulin = linep_capomulin.reset_index()
mouse = linep_capomulin.loc[linep_capomulin['Mouse ID'] == 's185']
mouse.sort_values(by=['Timepoint'])
plt.plot( mouse['Timepoint'], mouse['Tumor Volume (mm3)'])
plt.title('Tumor volume vs. time point for s185')
plt.xlabel('Time point')
plt.ylabel('Tumor volume')
plt.show()
```

```python
# Generate a scatter plot of average tumor volume vs. mouse weight for the Capomulin regimen

Avrg_volume_weight = df_merged.loc[df_merged['Drug Regimen'] == 'Capomulin']
Avrg_volume_weight1 = Avrg_volume_weight.groupby(['Mouse ID', 'Weight (g)']).mean()['Tumor Volume (mm3)'].reset_index()
plt.scatter(Avrg_volume_weight1['Weight (g)'], Avrg_volume_weight1['Tumor Volume (mm3)'])
plt.xlabel('Mouse Weight')
plt.ylabel('Average tumor volume')
plt.show()
```

## Correlation and Regression

```python
# Calculate the correlation coefficient and linear regression model 
# for mouse weight and average tumor volume for the Capomulin regimen

from scipy.stats import linregress

# Data For the plots
x_values = Avrg_volume_weight1['Weight (g)']
y_values = Avrg_volume_weight1['Tumor Volume (mm3)']

# Statistical values
(slope, intercept, rvalue, pvalue, stderr) = linregress(x_values, y_values)

# linear regression model
regress_values = x_values * slope + intercept
line_equation = 'y=' + str(round(slope,2)) + 'x +' + str(round(intercept,2))
# Plot and r value
plt.scatter(x_values, y_values)
plt.plot(x_values, regress_values, 'r-')
plt.annotate(line_equation,(6,10), fontsize = 15, color='red')
plt.xlabel('Mouse Weight')
plt.ylabel('Average tumor volume')
print(f'The correlation coefficient is {rvalue**2}')
plt.show()
line_equation
```
